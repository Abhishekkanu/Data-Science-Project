```python
# Handling Outliers
# Step 1: Box Plot for Numeric Columns
num.plot(kind="box", subplots=True, layout=(4, 5), figsize=(18, 12))

plt.suptitle("Box Plot for Numeric Columns")

plt.show()
```

```python
# Step 2: Identifying Outliers Using IQR
def Handling_Outlier(ColName):

  Q1 = np.quantile(ColName, 0.25)

  Q3 = np.quantile(ColName, 0.75)

  IQR = Q3-Q1

  Lower = Q1 - 1.5*IQR

  Upper = Q3 + 1.5*IQR

  outlier = ColName [ (ColName < Lower) | (ColName > Upper) ]

  return outlier
```

```python
adr_outliers = Handling_Outlier(num['lead_time'])

print(adr_outliers)
```

```python
# Purpose:
for i in num.columns:

  outliers = Handling_Outlier(data[i])

  outliers_percentage = (len(outliers) / len(data[i])) * 100

  print(f"{i}: {outliers_percentage:.2f}%")
```
